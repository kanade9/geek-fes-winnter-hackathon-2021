```python
def getSentence(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    texts = [s.getText().strip() for s in soup.find_all("p")]
    return texts
```

```python
d={}
for key in urls:
    retval = []
    for l in urls[key]:
        retval+=getSentence(l)
    d[key]=retval
```

```python
d
```

```python
def search(words):
    c = {}
    for key in d:
        c[key]=0
        for word in words:
            for s in d[key]:
                if word in s:
                    c[key]+=1
    return c
```

```python
search(["ウェブ","人工知能","DX"])
```

```python
model = Word2Vec.load("wiki.model")
```

```python
model.wv["人工知能"].shape
```

```python
def getVectors():
    c = {}
    for key in d:
        c[key]=np.zeros(200,)
        for s in d[key]:
            words = m.parse(s).strip().split(" ")
            for word in words:
                try:
                    c[key] += model.wv[word.lower()]
                except KeyError:
                    print(word)
    return c
```

```python
v = getVectors()
```

```python
v
```

```python
_v = [v[key] for key in v]
_v = np.array(_v)
```

```python
_v.shape
```

```python
pca = PCA()
v_pca = pca.fit_transform(_v)
```

```python
plt.figure(figsize=(12,12))
for i, label in enumerate(v.keys()):
    plt.scatter(v_pca[i,0],v_pca[i,1], label=label)
    plt.legend()
```
